```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(
    "FCB_Analysis_Datasets_FULL_SEASON.xlsx",
    sheet_name="match_logs"
)
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
plt.figure(figsize=(12,5))

plt.plot(df['Goals_For'])

plt.title("Bayern Goals Scored Per Match")
plt.xlabel("Match Number")
plt.ylabel("Goals Scored")

plt.show()
```

```python
plt.figure(figsize=(14,6))

plt.plot(
    df['Goals_For'],
    marker='o',
    linewidth=2
)

plt.title("FC Bayern Goals Scored Per Match", fontsize=16)

plt.xlabel("Match Number", fontsize=12)
plt.ylabel("Goals Scored", fontsize=12)

plt.grid(True)

plt.show()
```

```python
average_goals = df['Goals_For'].mean()

print("Average Goals Scored Per Match:", round(average_goals, 2))
```

```python
wins = (df['Result'] == 'W').sum()

total_matches = len(df)

win_percentage = (wins / total_matches) * 100

print("Win Percentage:", round(win_percentage, 2), "%")
```

```python
home_matches = df[df['Venue'] == 'Home']
away_matches = df[df['Venue'] == 'Away']

print("Home Wins:", (home_matches['Result'] == 'W').sum())
print("Away Wins:", (away_matches['Result'] == 'W').sum())
```

```python
venue_wins = [
    (home_matches['Result'] == 'W').sum(),
    (away_matches['Result'] == 'W').sum()
]

venues = ['Home Wins', 'Away Wins']

plt.figure(figsize=(6,5))

plt.bar(venues, venue_wins)

plt.title("Home vs Away Wins")

plt.ylabel("Number of Wins")

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.bar(
    venues,
    venue_wins
)

plt.title("FC Bayern Home vs Away Wins", fontsize=16)

plt.xlabel("Venue", fontsize=12)
plt.ylabel("Wins", fontsize=12)

plt.grid(axis='y')

plt.show()
```

```python
goals_scored = df['Goals_For'].sum()
goals_conceded = df['Goals_Against'].sum()

print("Total Goals Scored:", goals_scored)
print("Total Goals Conceded:", goals_conceded)
```

```python
categories = ['Goals Scored', 'Goals Conceded']
values = [goals_scored, goals_conceded]

plt.figure(figsize=(7,5))

plt.bar(categories, values)

plt.title("FC Bayern Attack vs Defense", fontsize=16)

plt.ylabel("Goals")

plt.grid(axis='y')

plt.show()
```

```python
average_possession = df['Possession'].mean()

print("Average Possession:", round(average_possession, 2), "%")
```

```python
plt.figure(figsize=(12,5))

plt.plot(
    df['Possession'],
    marker='o',
    linewidth=2
)

plt.title("FC Bayern Possession Per Match", fontsize=16)

plt.xlabel("Match Number")
plt.ylabel("Possession %")

plt.grid(True)

plt.show()
```

```python
shot_efficiency = goals_scored / df['Shots'].sum()

print("Shot Conversion Rate:", round(shot_efficiency * 100, 2), "%")
```

```python
result_counts = df['Result'].value_counts()

plt.figure(figsize=(7,5))

plt.bar(
    result_counts.index,
    result_counts.values
)

plt.title("FC Bayern Match Results", fontsize=16)

plt.xlabel("Result")
plt.ylabel("Number of Matches")

plt.grid(axis='y')

plt.show()
```

```python
attack_df = pd.read_excel(
    "FCB_Analysis_Datasets_FULL_SEASON.xlsx",
    sheet_name="attack_trio_stats"
)
```

```python
attack_df.head()
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    attack_df['Player'],
    attack_df['Goals']
)

plt.title("Attack Trio Goals", fontsize=16)

plt.xlabel("Player")
plt.ylabel("Goals")

plt.grid(axis='y')

plt.show()
```

```python
total_goals = attack_df.groupby('Player')['Goals'].sum()

print(total_goals)
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    total_goals.index,
    total_goals.values
)

plt.title("Total Goals by Bayern Attack Trio", fontsize=16)

plt.xlabel("Player")
plt.ylabel("Goals")

plt.grid(axis='y')

plt.show()
```

```python
total_assists = attack_df.groupby('Player')['Assists'].sum()

print(total_assists)
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    total_assists.index,
    total_assists.values
)

plt.title("Attack Trio Assists", fontsize=16)

plt.xlabel("Player")
plt.ylabel("Assists")

plt.grid(axis='y')

plt.show()
```

```python
gk_df = pd.read_excel(
    "FCB_Analysis_Datasets_FULL_SEASON.xlsx",
    sheet_name="goalkeeper_stats"
)

gk_df.head()
```

```python
total_clean_sheets = gk_df['Clean_Sheets'].sum()

print("Total Clean Sheets:", total_clean_sheets)
```
